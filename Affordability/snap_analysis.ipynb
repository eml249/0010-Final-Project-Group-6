```python
# Imports pandas and matplotlib for future reference
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 5]


# Sets the DataFrame to display ALL rows, regardless of size. 
# Done for clarity and easy comparison
pd.set_option('display.max_rows', None)

# Creates a table of all the neighborhoods used in this survey
# Uses only the percentage of residential SNAP holders in these neighborhoods to create a reflection
neighborhoods = pd.read_csv("Neighborhoods_with_SNAP_Data.csv")

# Reassigning the neighborhoods variable to only include the name of the neighborhood and the percentage of residents enrolled in SNAP
neighborhoods = neighborhoods[["SNAP_All_csv_Neighborhood", "SNAP_All_csv_Residential"]]

# Create a new variable making a dataframe from the previous variable
newNeighborhoods = pd.DataFrame(neighborhoods)

# Reassign the new variable to the revised version of the columns
# Renamed the columns for clarity 
newNeighborhoods = newNeighborhoods.rename(columns={
    'SNAP_All_csv_Neighborhood': 'Neighborhoods',
    'SNAP_All_csv_Residential': 'SNAP Enrollment'
})

# Sorts the neighborhoods alphabetically
sortedNeighborhoods = newNeighborhoods.sort_values('Neighborhoods')

# Replaces the index number with the name of the neigborhood for cleanliness and clarity
sortedNeighborhoods.set_index('Neighborhoods', inplace=True)

# Prints the table
sortedNeighborhoods
```

### Stay with me!
- While this table is *quite* large, with 90 neighborhoods being considered, I can still determine some useful information for Pittsburgh neighborhoods. Therefore, I am going to calculate and share the mean, median, maximum, and minimum of the 'SNAP Enrollment' column to give a better idea on how Bloomfield fares to all neighborhoods.

```python
# This replaces the percent sign with a blank, so that all the values in the column may be averaged out.
sortedNeighborhoods['SNAP Enrollment'] = sortedNeighborhoods['SNAP Enrollment'].replace('%', '', regex=True).astype(float)

# Mean variable storing the value of the average
mean = str(sortedNeighborhoods['SNAP Enrollment'].mean())

print("Mean (Average) SNAP enrollment for Pittsburgh neighborhoods: " + mean + "%\n")

# Setting up local variables
max = 0
min = 100

# Appends through each row in sortedNeighborhoods and finds the maximum value or minimum value
for index, row in sortedNeighborhoods.iterrows():
    potentialMax = row['SNAP Enrollment']
    potentialMin = row['SNAP Enrollment']
    if (max < potentialMax):
        max = potentialMax
        
    if (min > potentialMin):
        min = potentialMin
        
# max = sortedNeighborhoods.to_numpy().max()

print("Highest SNAP enrollment for Pittsburgh neighborhoods: ",max,"%\n")
print("Lowest SNAP enrollment for Pittsburgh neighborhoods: ",min,"%\n")

# Pandas function used to find the median of the 'SNAP Enrollment' column
median = str(sortedNeighborhoods['SNAP Enrollment'].median())

print("Median SNAP enrollment for Pittsburgh neighborhoods:",median,"%")
```

### Let's take a look at Bloomfield's metrics again to compare:

```python
for index, row in sortedNeighborhoods.iterrows():
    if (index == "Bloomfield"): 
        print(row)
```

### Therefore...

- Bloomfield has a higher SNAP enrollment compared to the mean enrollment (*48.216%*).

- Bloomfield has a higher SNAP enrollment compared to the median enrollment (*49.3%*).

- The average Pittsburgh neighborhood has a SNAP enrollment of around **50%**.

- Bloomfield's SNAP enrollment is at a staggering **58.1%**.

However, I believe this is still a valid argument in terms of affordability in Bloomfield. I would next like to provide a graph of the information I just presented to you. This may make for a better visual aid than the table/simple print statements provided previously.

```python
# Plots the data on a chart
sortedNeighborhoods['SNAP Enrollment'] = sortedNeighborhoods['SNAP Enrollment'].replace('%', '', regex=True).astype(float)

# Configuration of the actual graph for clarity and cleanliness
fig = sortedNeighborhoods.plot(
    kind="bar",
    title="SNAP Enrollment per neighborhood",
    color="yellow",
    edgecolor="blue"
)

# Configuration of the actual graph for clarity and cleanliness
fig.set(
    ylabel = "SNAP Enrollment (%)"
)

# Data points plotted for easier reading
plt.axhline(y = 48, color = 'b', linestyle = 'dashed', label="Mean") 
plt.axhline(y = 49.3, color = 'r', linestyle = '-', label="Median")
plt.plot(9, 58.1, color = 'black', marker = '*', markersize = 20, label="Bloomfield")


# Create max point and min points as well for additional information

# Minimum points
# Allegheny Center
plt.plot(0, 0, color = 'green', marker = '^', markersize = 15, label="Minimum")
# Central Business District
plt.plot(16, 0, color = 'green', marker = '^', markersize = 15)
# Chateau
plt.plot(21, 0, color = 'green', marker = '^', markersize = 15)
# North Shore
plt.plot(56, 0, color = 'green', marker = '^', markersize = 15)
# South Shore
plt.plot(70, 0, color = 'green', marker = '^', markersize = 15)
# Strip District
plt.plot(79, 0, color = 'green', marker = '^', markersize = 15)


# Maximum point
plt.plot(43, 94, color = 'purple', marker = 'X', markersize = 15, label="Maximum")

# Sets up a legend to preview the different markings on the graph
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'lower center') 

# Resizes the graph to better accommodate the many x-vals
fig = plt.figure(figsize=(10, 1))

# Command to show the graph I created
plt.show()
```
